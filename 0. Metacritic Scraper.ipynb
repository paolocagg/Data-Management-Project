```python
from selenium import webdriver
from bs4 import BeautifulSoup
from time import sleep
from pymongo import MongoClient
import pandas as pd
import json
```

```python
# Connect to mongodb
client = MongoClient('localhost', 27017)
# Create new database
db = client['Project2']
```

```python
op = webdriver.ChromeOptions()
# op.add_argument('headless')
driver = webdriver.Chrome()
```

# Scrape metacritic's PC reviews

```python
for i in range(786):
    top_games=[]
    driver.get("https://www.metacritic.com/browse/games/release-date/available/pc/date?page=" + str(i))
    soup = BeautifulSoup(driver.page_source, features="html.parser")
    for tr in soup.findAll('tr', class_=None):
        # initialize inner dictionary
        game = {}
        # append title
        game['title'] = tr.find("a", class_="title").find("h3").text
        # append link
        game['link'] = tr.find("a", class_="title")['href'].strip()
        # append metacritic score
        game['score'] = tr.find("a", class_="metascore_anchor").text.strip()
        # append release date
        game['relDate'] = tr.find("span", class_=None).text
        # append the newly created game dictionary to the top_games list
        top_games.append(game)
    metaPc.insert_many(top_games)
```

```python
i = 0
for game in metaPc.find():
    i += 1
    if game['score'] != "tbd":
        continue
    driver.get("https://www.metacritic.com" + game['link'] + "/critic-reviews")
    soup = BeautifulSoup(driver.page_source, features="html.parser")
    try:
        nRev = soup.find("span", class_="based").find_next_siblings()[0].text.strip().split(" ")[0]
    except:
        continue
    if int(nRev) == 0:
        continue
    ol = soup.find("ol", class_="reviews critic_reviews")
    grades = []
    for grade in ol.findAll("div", class_="review_grade"):
        grades.append(int(grade.text.strip()))
    metaPc.update_one({'_id': game['_id']}, {'$set' : {'score': int(sum(grades)/len(grades))}})
    print(game['link'] + ' updated')
```

# Upload to Database

```python
# Create new database collection
metaPc = db.metacriticPc
# upload to database
metaPc.insert_many(top_games)
```

```python
# Query the database to find how many documents have a reviews' score
metaPc.count_documents({"score": {"$ne": "tbd"}})
```

# Scrape all metacritic's reviews

```python
metaAll = db.metacriticAll
```

```python
diz = {'ps5': 17,
      'ps4': 89,
      'xbox-series-x': 13,
      'xboxone': 69,
      'switch': 93,
      'ios': 123,
      'stadia': 4}
```

```python
for key, value in diz.items():
    for i in range(value):
        top_AllGames = []
        # For every page get the HTML
        driver.get("https://www.metacritic.com/browse/games/release-date/available/" + key + "/date?page=" + str(i))
        soup = BeautifulSoup(driver.page_source, features="html.parser")
        # Find all of the reviews inside the page
        for tr in soup.findAll('tr', class_=None):
            # initialize inner dictionary
            game = {}
            # append title
            game['title'] = tr.find("a", class_="title").find("h3").text
            # append link
            game['link'] = tr.find("a", class_="title")['href'].strip()
            # append metacritic score
            game['score'] = tr.find("a", class_="metascore_anchor").text.strip()
            # append release date
            game['relDate'] = tr.find("span", class_=None).text
            # append the newly created game dictionary to the top_games list
            top_AllGames.append(game)
        metaAll.insert_many(top_AllGames)
```

```python
i = 0
for game in metaAll.find():
    i += 1
    if game['score'] != "tbd":
        continue
    driver.get("https://www.metacritic.com" + game['link'] + "/critic-reviews")
    soup = BeautifulSoup(driver.page_source, features="html.parser")
    try:
        nRev = soup.find("span", class_="based").find_next_siblings()[0].text.strip().split(" ")[0]
    except:
        continue
    if int(nRev) == 0:
        continue
    ol = soup.find("ol", class_="reviews critic_reviews")
    grades = []
    for grade in ol.findAll("div", class_="review_grade"):
        grades.append(int(grade.text.strip()))
    metaAll.update_one({'_id': game['_id']}, {'$set' : {'score': int(sum(grades)/len(grades))}})
    print(game['link'] + ' updated')
```
